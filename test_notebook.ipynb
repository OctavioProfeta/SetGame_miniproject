```python
from utilities import Card, find_sets, all_features_dict
import itertools
```

```python
# all possible cards in the game
all_cards = set()
for tuple in list(itertools.product(*all_features_dict.values())):
    all_cards.add(Card(tuple))

print("In total, there are {} different cards (3^4)".format(len(all_cards)))
print(all_cards)
```

```python
cards_in_round = [
    Card(('1', 'diamond', 'open', 'purple')), 
    Card(('2', 'diamond', 'open', 'purple')), 
    Card(('3', 'diamond', 'open', 'purple')),
    Card(('1', 'oval', 'open', 'purple')), 
    Card(('2', 'oval', 'open', 'purple')), 
    Card(('3', 'oval', 'open', 'purple'))
]

print(cards_in_round)
```

```python
sets = find_sets(cards_in_round)

if sets:
    print("Number of possible sets in this round: {}".format(len(sets)))
    print("The possible sets are: {}".format(sets))
else:
    print("There are no possible sets...")
```
